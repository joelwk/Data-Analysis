### Plot of histograms showing **Minutes Asleep**, **Minutes Awake**, **Number of Awakenings**, **Minutes Deep Sleep**

```python
hist_col = ['Minutes Asleep', 'Minutes Awake', 'Number of Awakenings', 'Minutes Deep Sleep']

plt.figure(figsize=(15, 15))
cnt=1
for i in hist_col:
    plt.subplot(4,2,cnt)
    sns.distplot(sleep[i], hist=True, color='darkblue', kde_kws=
    {'linewidth': 4})
    plt.title(i)
    cnt=cnt+1
plt.tight_layout(1)
plt.show()
```

### Finding correlations between columns by visualizing them through a Heatmap

```python
# Create the correalation
corr = sleep.corr()

#Plot the correlation
fig, ax = plt.subplots(figsize=(12,9))
sns.heatmap(corr, annot=True)
ax.set_xticklabels(ax.get_xticklabels(), rotation=20);
plt.show()
```
