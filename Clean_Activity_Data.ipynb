```python
import pandas as  pd
import numpy as np
```

```python
activity = pd.read_csv('Organized_Activity_Data.csv')
```

### Preview first 10 rows of data and display data types

```python
activity.head(10)
```

```python
activity.dtypes
```

### Converting the **Date** to datetime format

```python
activity['Date'] = pd.to_datetime(activity['Date'])
```

### We can iterate over the object columns to replace the commas and then iterate again to convert to int

```python
for i in ['Calories Burned', 'Steps', 'Minutes Sedentary', 'Activity Calories']:
    activity[i] = activity[i].replace(',','', regex=True)
```

```python
for i in ['Calories Burned', 'Steps', 'Minutes Sedentary', 'Activity Calories']:
    activity[i] = activity[i].astype(str).astype(int)
```
