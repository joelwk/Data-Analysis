```python
import numpy as np
import csv
import sys
import plotly.express as pxl
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn as sns
from sklearn.linear_model import LinearRegression
```

```python
activity = pd.read_csv('Activity Data.csv')
sleep = pd.read_csv('Sleep Data.csv')
```

```python
x1 = activity['Distance'].astype(str).astype(float)
y =  sleep['Minutes Asleep'].astype(str).astype(int)
```

```python
plt.scatter(x1,y)
plt.xlabel('Distance',fontsize=20)
plt.ylabel('Minutes Asleep',fontsize=20)
plt.show()
```

```python
x1 = x1.values.reshape(-1,1)
x = sm.add_constant(x1)
results = sm.OLS(y,x).fit()
results.summary()
```

```python
plt.scatter(x1,y)
yhat = x1*-3.8795+358.0465
fig = plt.plot(x1,yhat, lw=4, c='orange', label ='regression line')
plt.xlabel('Distance(miles)', fontsize = 20)
plt.ylabel('Minutes Asleep', fontsize = 20)
plt.show()
```
